```
#We define de ASCII_CHARACTERS
ASCII_CHARS = ["@", "#", "$", "%", "?", "*", "+", ";", ":", ",", "."]

#We test with Pikachu

path = "/content/2437349-pikachu.png"
image = PIL.Image.open(path)
```

```
#Function to rescale

def resize(image, new_width = 30):
    width, height = image.size
    new_height = new_width * height / width
    return image.resize((int(new_width), int(new_height)))

resize(image)    
```

```
#Function to convert the image to gray_scale

def to_greyscale(image):
    return image.convert("L") 

to_greyscale(resize(image))    
```

```
#Convert pixels to ascii

def pixel_to_ascii(image):
    pixels = image.getdata()
    ascii_str = "";
    for pixel in pixels:
        ascii_str += ASCII_CHARS[pixel//25];
    return ascii_str 

#Split the string based on width of the image

greyscale_image = to_greyscale(resize(image))  
ascii_str = pixel_to_ascii(greyscale_image)
img_width = greyscale_image.width
ascii_str_len = len(ascii_str)
ascii_img=""    

for i in range(0, ascii_str_len, img_width):
    ascii_img += ascii_str[i:i+img_width] + "\n"    
print(ascii_img)
```

```
#Put all into a big function

import PIL.Image
ASCII_CHARS = ["@", "#", "$", "%", "?", "*", "+", ";", ":", ",", "."]

def resize(image, new_width = 100):
    width, height = image.size
    new_height = new_width * height / width
    return image.resize((int(new_width), int(new_height)))

def to_greyscale(image):
    return image.convert("L")  

def pixel_to_ascii(image):
    pixels = image.getdata()
    ascii_str = "";
    for pixel in pixels:
        ascii_str += ASCII_CHARS[pixel//25];
    return ascii_str    


def draw_ascii():
    path = input("Enter the path to the image : \n")
    #path = "https://tse2.mm.bing.net/th?id=OIP.tg4xPiNytAGE3FdX_tfCmwHaG0&pid=Api"
    try:
        image = PIL.Image.open(path)
    except:
        print(path, "Unable to find image ")    
    
    #resize image
    image = resize(image);    
    
    #convert image to greyscale image
    greyscale_image = to_greyscale(image)    
    
    #convert greyscale image to ascii characters
    ascii_str = pixel_to_ascii(greyscale_image)
    
    #Split the string based on width of the image
    img_width = greyscale_image.width
    ascii_str_len = len(ascii_str)
    ascii_img=""    
    
    for i in range(0, ascii_str_len, img_width):
        ascii_img += ascii_str[i:i+img_width] + "\n"    
    print(ascii_img)

    #save the string to a file
    with open("ascii_image.txt", "w") as f:
        f.write(ascii_img);

draw_ascii()
```
